```python
import requests
import pandas as pd
import json
import time
```

```python
SUPERHERO_COUNT = 731
ACCESS_TOKEN = 'b0ac12ce3ed23a719bebd7ea8189bdf5'
all_heroes = []
```

```python
def get_hero(id):
    base_url = f"https://superheroapi.com/api/{ACCESS_TOKEN}"
    hero_url = f"{base_url}/{id}"
    print(f"Getting hero with ID {id} from {hero_url}")
    response = requests.get(hero_url)
  
    json_response = response.json()
    if json_response['response'] != "success":
        print(f"Hero with id {id} not found")
        print(json_response)
        return None
  
    return json_response
```

```python
for i in range(1, SUPERHERO_COUNT+1):
  hero = get_hero(i)
  all_heroes.append(hero)
  time.sleep(0.25)
```

```python
superheroes = pd.json_normalize(all_heroes)
superheroes
```

```python
superheroes.to_csv("superheroes.csv")
```
